```python
from smart_open import smart_open
from os import listdir
import numpy
import matplotlib.pyplot as plt
```

```python
eval_docs=[]
for file in listdir(path):
    if file.find("eval")!=-1:
        eval_docs.append(file)
print(eval_docs)
```

```python
def map_extractor(eval_file_path):
    map_array=[]
    open_file=smart_open(eval_file_path, 'rb', encoding='utf8')
    for line in open_file:
        if(line.split()[0]=="map") and (line.split()[1]!="all"):
            map_array.append(float(line.split()[2]))
    return map_array
```

```python
for doc in eval_docs:
    if doc.find("Only_TF_IDF")!=-1:
        map_Only_TF_IDF=map_extractor(path+doc)
    elif doc.find("Stemmer_BM25")!=-1:
        map_Stemmer_BM25=map_extractor(path+doc)
    elif doc.find("Stop_Stem_BM25")!=-1:
        map_Stop_Stem_BM25=map_extractor(path+doc)
    elif doc.find("Stop_Stem_TF_IDF")!=-1:
        map_Stop_Stem_TF_IDF=map_extractor(path+doc)
#print(np.arange(0, 51))
#plt.hist([map_Only_TF_IDF,map_Stemmer_BM25,map_Stop_Stem_BM25,map_Stop_Stem_TF_IDF])
#n, bins, patches = plt.hist(x=np.arange(0, 51) , bins=map_Stop_Stem_TF_IDF , range=(0, 50), color='#0504aa', alpha=0.7, rwidth=0.85)
#plt.xlabel('Topics')
#plt.ylabel('Average precision')
#plt.title('MAP Analysis')



bins = np.arange(1, 51)
width=0.25
fig1 = plt.figure(figsize=(15,5))
ax1 = fig1.add_subplot(1,1,1)
ax1.bar(bins, map_Only_TF_IDF, width, color='orange', label='Only_TF_IDF')
ax1.bar(bins+width, map_Stemmer_BM25, width, color='blue', label='Stemmer_BM25')
ax1.bar(bins+width*2, map_Stop_Stem_BM25, width, color='red', label='Stoplist_Stemmer_BM25')
ax1.bar(bins+width*3, map_Stop_Stem_TF_IDF, width, color='darkgreen', label='Stoplist_Stemmer_TF_IDF')
plt.xlabel('Topics')
plt.ylabel('Average precision')
plt.title('MAP Analysis')
```
